## **4. Hitung Entropy Awal**

$$
p_1 = \frac{7}{15}, \quad p_2 = \frac{8}{15}
$$

$$
Entropy(D) = -\frac{7}{15} \log_2 \frac{7}{15} - \frac{8}{15} \log_2 \frac{8}{15} \approx 0.996
$$

---

## **5. Entropy Setelah Split**

### **Group Left (≤ 4.0):**

Data No: 1, 4, 14, 2, 5, 11, 3, 12, 7, 13

* **10 data total**
* **7 Rosa gallica**
* **3 Rosa damascena**

$$
Entropy(left) = -\frac{7}{10}\log_2\frac{7}{10} - \frac{3}{10}\log_2\frac{3}{10} \approx 0.881
$$

### **Group Right (> 4.0):**

Data No: 9, 15, 10, 8, 6

* **5 data total**
* **5 Rosa damascena**

$$
Entropy(right) = -1 \cdot \log_2 1 = 0
$$

---

### **Information Gain:**

$$
IG = Entropy(D) - \left( \frac{10}{15} \cdot 0.881 + \frac{5}{15} \cdot 0 \right)
$$

$$
IG = 0.996 - 0.587 \approx \boxed{0.409}
$$

---

## **6. Gini Index**

### **Gini Awal:**

$$
Gini(D) = 1 - \left( \frac{7}{15} \right)^2 - \left( \frac{8}{15} \right)^2 \approx 0.497
$$

### **Gini Left (≤ 4.0):**

$$
Gini(left) = 1 - \left( \frac{7}{10} \right)^2 - \left( \frac{3}{10} \right)^2 = 1 - 0.49 - 0.09 = 0.42
$$

### **Gini Right (> 4.0):**

$$
Gini(right) = 1 - (1)^2 = 0
$$

### **Gini Split:**

$$
Gini_{split} = \frac{10}{15} \cdot 0.42 + \frac{5}{15} \cdot 0 = \boxed{0.28}
$$

---

## **Hasil Evaluasi untuk Fitur "Sepal Length":**

## Visualisasi Decision Tree

Kode ini membangun model pohon keputusan untuk mengklasifikasi bunga Rosa gallica dan Rosa damascena berdasarkan Sepal Length dan Sepal Width. Data dikodekan menggunakan LabelEncoder, model dilatih dengan DecisionTreeClassifier, lalu aturan pohon ditampilkan dengan export_text.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree, export_text
from sklearn.preprocessing import LabelEncoder

# === Data Baru ===
data = {
    "Sepal Length (cm)": [3.1, 3.4, 3.9, 3.2, 3.5, 4.9, 4.0, 4.7, 4.3, 4.6, 3.5, 3.9, 4.0, 3.2, 4.3],
    "Sepal Width (cm)": [0.5, 0.1, 0.7, 0.2, 0.6, 2.2, 2.5, 2.4, 2.8, 2.3, 0.2, 0.7, 2.4, 0.1, 2.3],
    "Kelas": [
        "Rosa gallica", "Rosa gallica", "Rosa gallica", "Rosa gallica", "Rosa gallica",
        "Rosa damascena", "Rosa damascena", "Rosa damascena", "Rosa damascena", "Rosa damascena",
        "Rosa gallica", "Rosa gallica", "Rosa damascena", "Rosa gallica", "Rosa damascena"
    ]
}
df = pd.DataFrame(data)

# === Label Encoding ===
le = LabelEncoder()
df['Label'] = le.fit_transform(df['Kelas'])  # gallica=0, damascena=1

# === Fitur dan Target ===
X = df[["Sepal Length (cm)", "Sepal Width (cm)"]]
y = df["Label"]

# === Train Decision Tree ===
clf = DecisionTreeClassifier(criterion="gini", max_depth=3, random_state=42)
clf.fit(X, y)

# === 1. Tampilkan Pohon Keputusan ===
print("\n=== Pohon Keputusan (Decision Tree) ===")
tree_rules = export_text(clf, feature_names=list(X.columns))
print(tree_rules)

```

## Visualisasi Splitting

Kode ini menguji dua data bunga baru menggunakan model pohon keputusan yang telah dilatih, lalu memvisualisasikan hasil klasifikasi dalam bentuk decision boundary. Dua bunga diuji berdasarkan nilai Sepal Length dan Sepal Width, dan hasil prediksi ditampilkan. Selanjutnya, kode menggambar batas klasifikasi di ruang fitur untuk menunjukkan area prediksi antara kelas Rosa gallica dan Rosa damascena.










```python
# === Uji 2 Data Baru ===
uji_baru = pd.DataFrame({
    'Sepal Length (cm)': [3.3, 4.5],
    'Sepal Width (cm)': [0.3, 2.1]
})

prediksi = clf.predict(uji_baru)
hasil = le.inverse_transform(prediksi)

print("\n=== Prediksi 2 Data Baru ===")
for i, (sl, sw) in enumerate(zip(uji_baru['Sepal Length (cm)'], uji_baru['Sepal Width (cm)'])):
    print(f"Data {i+1}: Sepal Length={sl}, Sepal Width={sw} → Prediksi: {hasil[i]}")

# === Visualisasi Decision Boundary ===
print("\n=== Visualisasi Decision Boundary ===")
plot_colors = "ry"
plot_step = 0.01

x_min, x_max = X['Sepal Length (cm)'].min() - 0.2, X['Sepal Length (cm)'].max() + 0.2
y_min, y_max = X['Sepal Width (cm)'].min() - 0.2, X['Sepal Width (cm)'].max() + 0.2

xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                     np.arange(y_min, y_max, plot_step))

grid = np.c_[xx.ravel(), yy.ravel()]
grid_df = pd.DataFrame(grid, columns=['Sepal Length (cm)', 'Sepal Width (cm)'])
Z = clf.predict(grid_df)
Z = Z.reshape(xx.shape)

plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, cmap=plt.cm.RdYlBu, alpha=0.6)

# Plot titik data training
for label, color in zip([0, 1], plot_colors):
    idx = np.where(y == label)
    plt.scatter(X.iloc[idx]['Sepal Length (cm)'], X.iloc[idx]['Sepal Width (cm)'],
                c=color, edgecolor='k', label=le.inverse_transform([label])[0])

plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.title("Decision Boundary (Sepal Length vs Sepal Width)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

Setelah prediksi dilakukan, kode memvisualisasikan decision boundary atau batas keputusan antara dua kelas bunga (Rosa gallica dan Rosa damascena) pada plot dua dimensi. ## Visualisasi Decision Tree dalam bentuk gambar

Kode ini menampilkan visualisasi struktur pohon keputusan (decision tree) yang telah dilatih sebelumnya. Visualisasi ini menunjukkan bagaimana fitur Sepal Length dan Sepal Width digunakan untuk membagi data dalam proses klasifikasi, lengkap dengan label kelas (Rosa gallica dan Rosa damascena), nilai Gini, jumlah sampel, dan distribusi kelas di setiap node.
```python
# === Visualisasi Struktur Pohon ===
plt.figure(figsize=(12, 3))
plot_tree(
    clf,
    feature_names=["Sepal Length (cm)", "Sepal Width (cm)"],
    class_names=le.classes_,  # gunakan label string hasil encoding
    filled=True,
    rounded=True,
    impurity=True  # tampilkan nilai Gini
)
plt.title("Decision Tree untuk Panthera tigris vs Panthera leo")
plt.tight_layout()
plt.show()
```
